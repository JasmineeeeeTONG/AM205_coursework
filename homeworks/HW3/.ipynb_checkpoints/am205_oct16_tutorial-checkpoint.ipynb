## Linear homogenous ODEs with constant coefficients

See: http://www.math.psu.edu/tseng/class/Math251/Notes-2nd%20order%20ODE%20pt1.pdf

Suppose we have an ODE of the form:

$a \ddot{y} + b \dot{y} + c y = 0$, for $a\ne 0$,

where $a$, $b$, and $c$ are constants.  This is a linear ODE with constant coefficients.  It is "homogeneous" becasuse the right hand side is zero.

Try substituting $y=e^{rt}$, $\dot{y} = r e^{rt}$, $\ddot{y} = r^2 e^{rt}$.

$a r^2 e^{rt} + b r e^{rt} + c e^{rt} = (a r^2 + b r + c) e^{rt} = 0 $.

Because $e^{rt} \ne 0$, solutions of the differential equation must satisfy $a r^2 + b r + c = 0$.  This equation is called the "characteristic equation."

There are three possible cases:

$b^2 - 4 a c > 0$: Two distinct, real roots

Solution will be of the form

$y = C_1 e^{r_1 t} + C_2 e^{r_2 t}$,

where $r_1$ and $r_2$ are the roots.

$b^2 - 4 a c < 0 $: Two complex conjugate roots

Remember Euler's formula: $e^{i\theta} = \cos\theta + i \sin\theta$.

The solution will be of the form

$y = C_1 e^{\lambda t} \cos \mu t + C_2 e^{\lambda t} \sin \mu t, $

where $r = \lambda \pm i \mu$.

$b^2 - 4 a c = 0$: One repeated real root

The solution will be of the form

$y = C_1 e^{r t} + C_2 t e^{r t}.$

If we have $y(0) = 1 $ and $\dot{y}(0) = 0$ can you solve for $C_1$ and $C_2$?

## Numerical integration of a system of ODEs

Last week we look at the harmonic oscillator (spring), set up a system of equations, numerically integrated them, and compared them to the analytic solution.

Today let's look at a system that is closer to the "asteroid problem" in HW3.  In particular, we will look at the two-body problem, a planet orbiting a star.

The equations of motion are given by Newton's law of universal gravitation:

$
\begin{equation}
\ddot{\vec{r}} = -\left(\frac{GM}{r^3}\right) \vec{r}
\end{equation}
$

In 2-D, this can be written as a system of four coupled equations:

$
\begin{eqnarray}
\dot{x} &= v_x\\
\dot{y} &= v_y\\
\dot{v_x} &= -\left(\frac{GM}{r^3}\right) x\\
\dot{v_y} &= -\left(\frac{GM}{r^3}\right) y\\
\end{eqnarray}
$

```python
def keplerian(state, t, GM):
    x, y, vx, vy = state
    r2 = x*x + y*y
    r = np.sqrt(r2)
    #f = -GM/(r2*r)*np.exp(-0.01*t)
    f = -GM/(r2*r)
    return [vx, vy, f*x, f*y]
```

Let's start our planet on the x-axis at x=1, moving perpendicular to the x-axis with unit velocity.

```python
t = np.linspace(0, 7, 101)
GM = 1.0
state0 = [1.0, 0.0, 0.0, 1.0]
numerical_sol = sp.integrate.odeint(keplerian, state0, t, args=(GM,))

x = numerical_sol[:,0]
y = numerical_sol[:,1]
```

```python
fig=plt.figure(figsize=(12, 12))
plt.plot(x, y)
plt.plot([0], [0], marker='o', markersize=5, color="red")
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)
```

The resulting orbit is circular.

```python
plt.plot(t, x)
plt.plot(t, y)
```

And we can check the total energy (per unit mass of the orbiting body), which is the sum of the kinetic energy and potential energy:

$E = \frac{1}{2} v^2 - \frac{GM}{r}$

```python
def energy(state, GM):
    x = state[:, 0]
    y = state[:, 1]
    vx = state[:, 2]
    vy = state[:, 3]
    r2 = x*x + y*y
    r = np.sqrt(r2)
    return 0.5*(vx*vx + vy*vy) - GM/r
```

```python
ener = energy(numerical_sol, GM)
plt.plot(t, (ener-ener[0])/ener[0])
```

Now let's give our planet a little more speed.  The orbit becomes elliptical.

There are many different ways to numerically integrate systems of equations.  Specialized techniques often exploit the properties of the systems they are designed to solve.  Here's a simple example that we will not discuss in section but which might be interesting.

```python
def kick(state, GM, dt):
    x, y, vx, vy = state
    r2 = x*x + y*y
    r = np.sqrt(r2)
    f = -GM/(r2*r)
    return [x, y, vx + dt*f*x, vy + dt*f*y]
    
def drift(state, dt):
    x, y, vx, vy = state
    return [x + dt*vx, y + dt*vy, vx, vy]
```

```python
def symplectic_1st(state, GM, dt):
    statep = drift(state, dt)
    staten = kick(statep, GM, dt)
    return staten
```

```python
GM = 1.0
state0 = [1.0, 0.0, 0.0, 1.0]
numerical_sol = np.zeros((201, 4))
t = np.zeros((201))
numerical_sol[0] = state0
t[0] = 0.0
dt = 0.05
for i in range(1, 201):
    state = symplectic_1st(state0, GM, dt)
    numerical_sol[i] = state
    t[i] = t[i-1] + dt
    state0 = state
    
x = numerical_sol[:,0]
y = numerical_sol[:,1]
```

```python
fig=plt.figure(figsize=(12, 12))
plt.plot(x, y)
plt.plot([0], [0], marker='o', markersize=5, color="red")

plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)

```

```python
ener = energy(numerical_sol, GM)
plt.plot(t, (ener-ener[0])/ener[0])
```

What?  The energy looks oscillatory.  What's up with that?  Let's look over a longer interval.

```python
GM = 1.0
state0 = [1.0, 0.0, 0.0, 1.0]
nsteps=2000
numerical_sol = np.zeros((nsteps+1, 4))
t = np.zeros((nsteps+1))
numerical_sol[0] = state0
t[0] = 0.0
dt = 0.05
for i in range(1, nsteps+1):
    state = symplectic_1st(state0, GM, dt)
    numerical_sol[i] = state
    t[i] = t[i-1] + dt
    state0 = state

x = numerical_sol[:,0]
y = numerical_sol[:,1]
```

```python
ener = energy(numerical_sol, GM)
plt.plot(t, (ener-ener[0])/ener[0])
```

Yes, it keeps on oscillating.  This what's called a first order 'symplectic' integrator.  First order meaning that the amplitude of the energy oscillation is proportional to the step size.

We can make a second order version, like so:

```python
def symplectic_2nd(state, GM, dt):
    statep = drift(state, 0.5*dt)
    staten = kick(statep, GM, dt)
    statef = drift(staten, 0.5*dt)
    return statef
```

```python
GM = 1.0
state0 = [1.0, 0.0, 0.0, 1.0]
nsteps=2000
numerical_sol = np.zeros((nsteps+1, 4))
t = np.zeros((nsteps+1))
numerical_sol[0] = state0
t[0] = 0.0
dt = 0.05
for i in range(1, nsteps+1):
    state = symplectic_2nd(state0, GM, dt)
    numerical_sol[i] = state
    t[i] = t[i-1] + dt
    state0 = state
    
x = numerical_sol[:,0]
y = numerical_sol[:,1]
```

```python
fig=plt.figure(figsize=(12, 12))
plt.plot(x, y)
plt.xlim()
plt.plot([0], [0], marker='o', markersize=5, color="red")

plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)
```

And the energy oscillation amplitude is much smaller.

```python
ener = energy(numerical_sol, GM)
plt.plot(t, (ener-ener[0])/ener[0])
```

```python
GM = 1.0
state0 = [1.0, 0.0, 0.0, 1.4]
nsteps=200000
numerical_sol = np.zeros((nsteps+1, 4))
t = np.zeros((nsteps+1))
numerical_sol[0] = state0
t[0] = 0.0
dt = 0.005
for i in range(1, nsteps+1):
    state = symplectic_2nd(state0, GM, dt)
    numerical_sol[i] = state
    t[i] = t[i-1] + dt
    state0 = state
    
x = numerical_sol[:,0]
y = numerical_sol[:,1]
```
